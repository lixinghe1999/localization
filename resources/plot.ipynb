```python
import matplotlib.pyplot as plt
```

```python
esc50_acc = [10, 36, 61, 78, 83, 92]
audio_length = [0.5, 1, 2, 3, 4, 5]
fig, ax = plt.subplots(1, 1)
ax.bar(range(len(esc50_acc)), esc50_acc)
ax.set_ylabel('ESC-50 Accuracy %')
ax2 = ax.twinx()
ax2.plot(audio_length, color='r')
plt.show()
```
